## Overplot the data

```python
import h5py
```

```python
f = h5py.File("../data/reduced/GW_Ori_IGRINS_H_20161119.hdf5", "r")

#list(f.items())
wvl_H_20161119 = np.array(f['wls'])
flx_H_20161119 = np.array(f['fls'])
sig_H_20161119 = np.array(f['sigmas'])

f.close()
```

```python
f = h5py.File("../data/reduced/GW_Ori_IGRINS_H_20161009.hdf5", "r")

wvl_H_20161009 = np.array(f['wls'])
flx_H_20161009 = np.array(f['fls'])
sig_H_20161009 = np.array(f['sigmas'])

f.close()
```

```python
plt.figure(figsize=(14,8))
o=17
plt.step(wvl_H_20161009[o,:], flx_H_20161009[o,:]/np.nanmedian(flx_H_20161009[o,:]), label='GW Ori 20161009')
plt.step(wvl_H_20161119[o,:], flx_H_20161119[o,:]/np.nanmedian(flx_H_20161119[o,:]), label='GW Ori 20161119')

plt.legend(loc='best')

plt.ylim(ymin=0)
plt.xlabel("$\lambda$ ($\mu$m)")
plt.ylabel("Raw signal (ADU)");
plt.xlim(np.min(wvl_H_20161009[o,:])+25.0, np.max(wvl_H_20161009[o,:]))
plt.ylim(0.8, 1.5)
```
